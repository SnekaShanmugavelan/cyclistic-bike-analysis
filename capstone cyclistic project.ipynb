```python
import pandas as pd
import numpy as np
```

```python
df_2019 = pd.read_csv("Divvy_Trips_2019_Q1.csv")
df_2020 = pd.read_csv("Divvy_Trips_2020_Q1.csv")
```

```python
df_2019.head()
df_2020.head()
```

```python
df_2019.info()
df_2020.info()
```

```python
df_2019 = df_2019.rename(columns={"trip_id" : "ride_id", 
                                  "start_time" : "started_at", 
                                  "end_time" : "ended_at", 
                                  "usertype" : "member_casual"})
```

```python
df_2020 = df_2020[["ride_id",
                  "started_at", "ended_at",
                  "member_casual"]]
```

```python
df = pd.concat([df_2019, df_2020], ignore_index = True)
```

```python
df = df.dropna()
```

```python
df["started_at"] = 
pd.to_datetime(df["started_at"]) 
df["ended_at"] = 
pd.to_datetime(df["ended_at"])
```

```python
df["started_at"] = pd.to_datetime(df["started_at"])
df["ended_at"] = pd.to_datetime(df["ended_at"])
```

```python
df["ride_length_minutes"] = (df["ended_at"] - df["started_at"]).dt.total_seconds() / 60
```

```python
df["day_of_week"] = df["started_at"].dt.day_name()
```

```python
np.mean(df["ride_length_minutes"])
np.max(df["ride_length_minutes"])
np.min(df["ride_length_minutes"])
```

```python
df["member_casual"].value_counts()
```

```python
df.groupby("member_casual")
["ride_length_minutes"].mean()
```

```python
df.groupby("member_casual")["ride_length_minutes"].mean()
```

```python
df.groupby(["member_casual", "day_of_week"]).size()
```

```python
df.groupby(["member_casual", "day_of_week"])["ride_length_minutes"].mean()
```

```python
def ride_summary(data):
    return {
        "avg_ride": data["ride_length_minutes"].mean(),
        "max_ride": data["ride_length_minutes"].max(),
        "total_rides": len(data)
    }

ride_summary(df)
```

```python
df.to_csv("cyclistic_cleaned.csv", index=False)
```
